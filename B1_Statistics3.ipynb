# 2) 2-Sample T-Test(138pp)

## 가설 설정:
- H0: A와 B 공급자 간에 차이가 없다.
- H1: A와 B 공급자 간에 차이가 있다.

```python
#데이터
df1 = pd.DataFrame({'A': [1.883, 1.715, 1.799, 1.768, 1.711, 1.832, 1.427, 1.344]})
df2 = pd.DataFrame({'B': [1.435, 1.572, 1.486, 1.511, 1.457, 1.548, 1.404, 1.883]})
```

```python
df3 = df1.append(df2)
df3.describe() #데이터 확인
```

A공급자와 B공급자의 제품 수는 8개이고 평균은 1.68, 1.53으로 공급자 간 해당 평균의 차이가 있는지 확인하고자 함

## 유의수준 0.05로 2-Sample t-test 검증:

```python
t_result = stats.ttest_ind(df1, df2)
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
```

```python
print("2-Sample t-test\n t 통계량 :{} \n p-value :{}".format(t, p))
```

## 결론: 공급자 A와 공급자 B의 제품 간의 평균 차이는 없다.(p > 0.05)

# 4) 1 Proportion Test(148pp)

## 가설 설정:
- H0 : 앞면이 나올 확률은 0.5이다.
- H1 : 앞면이 나올 확률은 0.5가 아니다.

## 유의수준 0.05에서 가설 검정:

```python
count = 40
n_obs = 100
value = 0.5
stat, pval = proportions_ztest(count, n_obs, value)
```

```python
print("1 Proportion test")
print("z: {0:0.3f}".format(stat))
print("p: {0:0.3f}".format(pval))
```

## 결론: 100번 던지기에서 앞면이 0.4확률로 나온 동전은 정상적이지 않다(p<0.05)

# 5) 2 Proportion Test(151pp)

## 가설 설정:
- H0 : 두 공정의 불량률은 같다.
- H1: 두 공정의 불량률은 다르다.

## 유의 수준 0.05에서 가설 검증:

```python
count = np.array([14, 5])
nobs = np.array([1200, 1200])
```

```python
e1, e2 = count[0]/nobs[0] , count[1]/nobs[1]
```

```python
print(e1.round(4), e2.round(4))
```

공장 A의 불량률은 0.0117 / 공장 B의 불량률은 0.0042 이며 해당 불량률이 차이가 있는지를 검증함

```python
stat, pval = proportions_ztest(count, nobs)
```

```python
print("2- Proportion test")
print("z: {0:0.3f}".format(stat))
print("p: {0:0.3f}".format(pval))
```

## 결론 : 공장 A와 공장 B의 불량률은 다르다(p<0.05)

## 데이터 분석: 
- 강종별과 유형별 모두 고려하여 시간의 차이가 있는지를 확인한다면 two-way ANOVA가 적합하나, 우선 강종별 one-way ANOVA, 유형별 one-way ANOVA를 분석하겠습니다.

```python
#강종별 one-way anova
f_result = stats.f_oneway(wheel['time'][wheel['gangjong']=='a'], wheel['time'][wheel['gangjong']=='b'], wheel['time'][wheel['gangjong']=='c'], wheel['time'][wheel['gangjong']=='d'])
```

```python
f, p = f_result.statistic.round(3), f_result.pvalue.round(3)
```

```python
print("One-way 분석(강종별 시간의 차이) :\n F통계량: {} \n p-value : {}".format(f, p))
```

## 결론 1: 강종별 2차 정련의 시간에는 차이가 나타납니다(p<0.05)

```python
#유형별 one-way ANOVA
f_result = stats.f_oneway(wheel['time'][wheel['type']=='a'], wheel['time'][wheel['type']=='b'])
```

```python
f, p = f_result.statistic.round(3), f_result.pvalue.round(3)
```

```python
print("One-way 분석(강종별 시간의 차이) :\n F통계량: {} \n p-value : {}".format(f, p))
```

## 결론 2: 유형별 2차 정련의 시간에는 차이가 나타난다(p<0.05)

## 추가분석: Two-way ANOVA
### 강종과 타입별 2가지 조건을 포함하여 2차정련 시간의 차이가 있는지를 분석

```python
#sm.stats.anova_lm(model, type=2)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('time ~ C(type) + C(gangjong) + C(type):C(gangjong)', data=wheel).fit()
#sm.stats.anova_lm(model)
print(sm.stats.anova_lm(model, type=5))
```

## two-way anova 결론: 강종과 type에 따라서 시간의 차이가 달라지지는 않습니다. 예를 들어 강종이 b일 때는 type a과 type b의 시간의 차이가 나타난다던가 하지는 않는다고 보여집니다(p > 0.05)
- 주효과 : 강종별 시간의 차이는 유의미함(p<0.05), 
- 주효과 : type별 시간의 차이는 유의미하지 않음(p>0.05). 
   - 이 부분은 위의 one-way와 F값(F통계량: 9.654 , p = 0.005)과 다른데 왜 다른지는 좀 더 생각해보겠습니다!!
- 상호작용 확인 : 강종과 type을 모두 고려했을 때의 시간의 차이는 나타나지 않는다(p>0.05)


# 7) 단순선형회귀(204pp)

- H1: 초코칩의 강도와 굽는 시간은 인과 관계가 있다. B1은 0이 아니다.

### 변수 간 경향성 파악:

```python
choco_raw.plot.scatter('strength', 'BakingTime')
```

#### 일부 데이터를 제외하면 강도와 굽는 시간의 상관관계가 나타나는 것처럼 보임

강도와 굽는 시간 간 0.804의 상관 계수로 높은 선형 관계를 보인다

### 회귀 모델 생성:

```python
reg_model=smf.ols(formula ="strength ~ BakingTime", data=choco_raw)
reg_result=reg_model.fit()
print(reg_result.summary())
```

### 회귀 모형 오차의 기본가정 검정

- 등분산성 검정

```python
plt.scatter(reg_result.fittedvalues, reg_result.resid)
plt.xlabel("fitted value")
plt.ylabel("residual")
plt.axhline(y=0, linestyle="--")
```

잔차의 등분산성을 그래프로 확인하였을 때, 등분산성을 만족합니다
- 잔차의 독립성 검정

```python
plt.plot(choco_raw["BakingTime"], reg_result.resid, marker="o")
plt.xlabel("Data")
plt.ylabel("residual")
plt.axhline(y=0, linestyle="-")
plt.axhline(y=3*np.std(reg_result.resid), linestyle="--")
plt.axhline(y=-3*np.std(reg_result.resid), linestyle="--")
```

- 정규성 검정

```python
qqplot(reg_result.resid, line="q")
```

```python
from statsmodels.stats.stattools import omni_normtest
omni_normtest(reg_result.resid, axis=0)
```

### 정규 분포 직선을 중심으로 분포하며 pvalue>0.05이므로 정규 분포를 따릅니다. 

## 결론 : 초코칩의 강도 = 30.87*굽는 시간 + 13.00 
## 굽는 시간이 1단위 증가할 수록 초코칩의 강도는 30정도 증가한다고 할 수 있다.
