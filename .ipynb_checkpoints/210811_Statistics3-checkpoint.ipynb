# 1 sample t-test

고객만족도 평균은 76.7
- 10개의 고객만족도 데이터
- 개선 활동이 만족도 변화를 시켰는지
- 유의수준 알파값(a) = 0.05

DATA 수집

```python
df = pd.DataFrame({'sample': [74.5, 81.2, 73.8, 82.0, 76.3, 75.7, 80.2, 72.6, 77.9, 82.8]})
```

정규성 검정

```python
shapiro(df) #정규성 검정도 해주셔야 합니다.
```

정규성을 만족한다

one-sample t-test

```python
t_result = stats.ttest_1samp(df, 76.7)
```

결과는 t_result에 저장되어 있음

```python
t_result
```

```python
# 1-Sample t-test의 t값 p값 저장
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
```

```python
print("1-Sample t-test")
print("t통계량 : {:}".format(t))
print("p-value: {}".format(p))
```

# 2-sample t-test

독립적인 두 집단 평균 비교
- a 집단 점수와 b집단의 설문조사 차이
- 유의수준은 0.05 
- 데이터는 10점 척도이며 n = 20임

데이터 수집

```python
df1 = pd.DataFrame({'sample': [6, 5, 5, 4, 6, 7, 6, 4, 5, 6, 4, 5, 5, 6, 4, 8, 6, 5, 6, 7]})
df2 = pd.DataFrame({'sample': [7, 5, 7, 8, 7, 8, 8, 5, 7, 6, 5, 5, 6, 6, 5, 7, 9, 7, 7, 8]})
```

정규성 검정

```python
# 정규성 검정
shapiro(df1)
```

```python
shapiro(df2)
```

등분산성 검정

```python
stats.levene(df1['sample'],df2['sample'])
```

```python
stats.bartlett(df1['sample'], df2['sample'])
```

two-sample t-test

```python
t_result= stats.ttest_ind(df1, df2)
```

```python
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
```

```python
print("2-Sample t-test")
print("t검정통계량: {}".format(t))
print("p-value: {}".format(p))
```

# 1 Proportion test
- 한 집단의 비율이 특정 비율과 같은지를 검정하는 도구

고급강 a제품을 가공하는 국내 고객사는 전체 중에 10% 정도이다.한 해동안 특정 제품 품질에 대해 노력하여 전체고객 중 100여 개의 업체를 표본으로 한 결과 15개의 업체가 만족을 표현. 
- 품질 개선을 한 결과로 기존보다 전체 고객 중 사용비율의 차이가 있는가 
- 유의수준 = 0.05
- 가설: p = 0.1 / h1 : p 는 0.1이 아니다

```python
count = 15
n_obs = 100
value = 0.1

stat, pval = proportions_ztest(count, n_obs, value)
```

```python
print("1 Proportion test")
print("z: {0:0.3f}".format(stat))
print("p: {0:0.3f}".format(pval))
```

특정 제품에 대해 품질 개선의 결과 기존 대비 사용비율이 차이가 있다고 할 수 없다

```python
stat1, pval1 = proportions_ztest(150, 1000, 0.1)
print("1 Proportion test")
print("z: {0:0.3f}".format(stat1))
print("p: {0:0.3f}".format(pval1))
```

```python
print("1 Proportion test")
print("z: {0:0.3f}".format(stat1))
print("p: {0:0.3f}".format(pval1))
```

# 카이스퀘어 검정

- 카이제곱 검정은 관찰된 빈도가 기대되는 빈도와 의미 있게 다른지의 여부를 검증하는 검증방법이다
- 자료의 빈도
- 특히, 범주형 자료 분석에 이용
- 예) 2개 공장에서 생산되는 제품의 품질의 차이를 확인하기 위해
- 1공장 200, 2공장 250 제품 수집
- 1공장 1등급 100, 2등급 60, 3등급 40
- 2공장 1등급 150, 2등급 70, 3등급 30
- 두 공장의 품질 수준이 같은지?

생산제품의 규격 3종류
- 근무조별 제품 생산 부하의 차이가 있는지
- 유의수준 0.05

```python
df = pd.DataFrame({'HSG': [270, 260,236, 234], 'SS':[228, 285,225,262], 'SPA':[277, 284, 231, 208]})
```

```python
chi, pval, dof, expected = stats.chi2_contingency(df)
```

```python
print("chi-square test")
print("chisq: {0:0.3f}".format(chi))
print("p: {0:0.3f}".format(pval))
print("degree of freedom: {}".format(dof))
print('expected value: \n{}'.format(expected.round(3)))
```

유의 수준 5%에서 근무조별로 제품을 생산하는 부하 차이가 있다고 할 수 있다

노트북 컴퓨터 6가지 제품에 대해 이미지를 중복 선택할 수 있도록 32명의 소비자를 대상으로 설문 조사한 표이다. 각 제품에 대해 이미지 차이가 있는지
- 유의수준 0.05
- a~f : 제품명
- 7개 행 : a/s, 가격, 성능, 확장성, 디자인, 안정성, 기능성

# one way Anova

3집단의 토익 성적 비교
- 유의수준 0.05

```python
df = pd.DataFrame({'A':[892, 623, 721, 678, 723, 790, 720, 670, 690, 771], 'B':[721, 821, 910, 678, 723, 790, 711, 790, 745, 891], 'C':[621, 915, 888, 721, 894, 834, 841, 912, 845, 889]})
```

```python
df.boxplot(['A', 'B','C'])
```

```python
f_result=stats.f_oneway(df['A'], df['B'], df['C'])
```

```python
f, p = f_result.statistic.round(3), f_result.pvalue.round(3)
```

```python
print("One-way")
print("F통계량 : {}".format(f))
print("p-value: {}".format(p))
```

유의수준 5%에서 검정결과 p값 < 0.05 이므로 그룹 간의 평균 차이가 있다고 할 수 있다

# 단순선형 회귀분석

- 독립변수가 종속변수에 미치는 영향력의 크기를 측정하여
- 독립변수의 일정한 값에 대응되는 종속변수의 값을 예측하기 위한 통계적 분석방법
- 영국 유전학자 갈톤(Galton)에 의해 처음 정의 된 용어, 회귀

- 잔차항의 필요조건
     - 잔차항은 정규분포를 따른다
     - 잔차항의 평균은 0
     - 잔차항의 분산은 모분산이고 등분산
     - 잔차항은 서로 독립이다
- 잔차의 전제사항 1. 정규성, 2.등분산성, 3.동립성임
- 다중공선성 확인 VIF 분산팽창계수 > 10 이면 다중공선성이 존재한다고 볼 수 있음.
- 다중공선성이란 독립변수들 사이에 상관관계를 갖고 있는 현상

분석 절차:
- 데이터 구성하기
- 변수 간의 경향성 파악 - 그래프 분석
- 회귀 모델 생성
     - 모델의 적합성 검토
         - 분산분석 / 결정계수 확인
     - 회귀계수 유의성 검토
- 오차의 기본가정 검토
- 결론 도출

### 회귀 모델 생성

```python
reg_model=smf.ols(formula="MINUTES ~ UNITS", data = df_raw)
reg_result = reg_model.fit()
print(reg_result.summary())
```

분석 자료 수 observation 14개, df residuals 잔차 자유도 12, df model 모델 자유도 1
- 분산분석 결과: p값이 유의 수준 0.05보다 적으므로 회귀모델로서의 유의함
- 설명력 r-squared (r^2) : 0.987 -> units로 minutes의 변동의 98.7%을 설명할 수 있음
- 회귀계수 검정:
    - 절편의 유의성은 검토하지 않음.(회귀식에는 절편의 유의성 상관없이 절편 포함
    - units의 p값이 유의수준 0.05보다 작으므로 유의
- omnibus, prob(omnibus)/jarque-bera(jb)는 잔차의 정규성을 뜻함
