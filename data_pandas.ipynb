# filter

```python
df = pd.DataFrame({"ABC": ["A","B","C","A"], 
    "abc":["a","b","c","c"],
    "group1":["G1","G2","G3","G4"],
    "value1":[11,13,12,14],
    "val1_mod":[0.231,0.197,0.213, 0.199] })
```

```python
df
```

```python
df.filter(items=['group1', 'value1'])
```

```python
# regex(regular express)
df.filter(regex='2', axis=0) #2가 포함된 행
```

```python
df.filter(regex='d$', axis=1) # d가 포함된 열
```

```python
df.filter(like='a', axis=1)
```

```python
df.filter(regex='[Aa]', axis=1)
```

# concat

```python
df1 = pd.DataFrame({
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]},
        index=[0, 1, 2, 3])

df2 = pd.DataFrame({
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"]},
        index=[4, 5, 6, 7])

df3 = pd.DataFrame({
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"]},
        index=[8, 9, 10, 11])
```

```python
pd.concat([df1, df2, df3]) #세로 결합
```

```python
df1.append([df2, df3])
```

```python
pd.concat([df1, df2, df3], axis=1) # 가로 결합
```

```python
result =pd.concat([df1, df2, df3], axis=1, ignore_index=True)  # 인덱스 새롭게 만들어짐 
result
```

```python
result.columns = ["Col_" + str(i) for i in range(12)] #인덱스 이름 만들어주기
```

```python
result
```

# merge

```python
left = pd.DataFrame({
        "key1": ["K0", "K1", "K2", "K3"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame({
        "key1": ["K0", "K1", "K2", "K3"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]})
```

```python
display(left, right)
```

```python
pd.merge(left, right, on=['key1', 'key2']) #innner default
```

```python
pd.merge(left, right, how='left')
```

```python
pd.merge(left, right, how='right')
```

```python
pd.merge(left, right, how='outer')
```

```python
pd.merge(left, right, how='outer', indicator=True) #기준치를 표시
```

# join

```python
left = pd.DataFrame({
    "key1": [1, 2, 3, 4],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame({
    "C": ["C1", "C2", "C5"],
    "D": ["D1", "D2", "D5"]},
    index=[1, 2, 5])
```

```python
display(left, right)
```

```python
left.join(right)
```

```python
left.join(right, how='right')
```

```python
left.join(right, how='inner')
```

```python
left.join(right, how='outer')
```

```python
left.join(right, on='key1')
```

# join과 merge의 비교

```python
pd.merge(left, right, left_on='key1', how='left', right_index=True)
```

# crosstab

```python
df = pd.DataFrame({'factor1': ['A','A','A','A','A','B','B','B','B'],
                  'factor2':['a','b','c','a','b','a','b','c','c'],
                  'value':[11,13,12,12,14,12,13,11,12]})
```

```python
df
```

```python
#df 데이터 생성
pd.crosstab(df.factor1, df.factor2)
```

```python
#행/열 명칭 부여 및 margin 추가
pd.crosstab(df.factor1, df.factor2,
           rownames=['Fact1'], colnames=['Fact2'],
           margins=True)
```

```python
#index 1개, columns 복수 개 컬럼 지정 가능
pd.crosstab(df.factor1, [df.factor2, df.value])
```

```python
#각 셀의 빈도 기준 정규화 확률 산출
pd.crosstab(df.factor1, df.factor2, margins=True,
           normalize=True)
```
