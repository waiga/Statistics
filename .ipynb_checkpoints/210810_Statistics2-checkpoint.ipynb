# Statistics

확률 계산
- 확률밀도: x 값에서의 확률밀도함수값 f(x)
- 누적확률(Cumulative probability) : 음수 무한대에서 x 값까지의 확률밀도 누적 값
- 역 누적확률(inverse cumulative probability): P(X<= ?) = Y를 만족하는 x값

```python
import numpy as np
import scipy
from scipy import stats
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportions_ztest
```

```python
# 누적 확률 값 계산하는 걸 함수로 한 번 만들어봤음
def cumulativeProb(x, mu, sigma):
    prob = stats.norm.cdf(x, mu, sigma)
    return ((1-prob)*100)
```

```python
cumulativeProb(135, 115, 20)
```

```python
# 누적 확률 값 계산: 평균115분, 표준편차 20분 소요. 135분 이상 걸리는 작업비율은?
mu = 115
sigma = 20
x = 135
prob = stats.norm.cdf(x, mu, sigma)
print("{0}분 이상 걸리는 A/S 작업 비율: {1:.1f}%".format(x, (1-prob)*100))
```

```python
#평균 175, 표준편차 5 인 키에서 180 cm인 경우 몇 %의 범위인지
prob = stats.norm.cdf(180, 175, 5)
print("{0}cm 이상인 키의 비율은: {1:.1f}%".format(180, (1-prob)*100))
```

```python
mu = 12
sigma = 3
x1 =15 

prob1 = stats.norm.cdf(x1, mu,sigma)
print("{0}km 이상 {1}km 이하를 달릴 확률은 {2:.1f}% 에 해당함".format(mu,x1,(prob1-0.5)*100))
```

```python
# 주행거리가 평균 12, 표준편차 3 인 정규분포일 때 
# 9km 이상 15km 이하를 달릴 확률은?
mu = 12
sigma = 3
# 9km까지의 확률
km_9 = stats.norm.cdf(9, 12, 3)
# 15km까지의 확률
km_15 = stats.norm.cdf(15, 12, 3)
# 9km 이상 15km 이하를 달릴 확률은
print("9km이상 15km 이하를 달릴 확률은 {0:.1f}%".format((km_15 - km_9)*100))
```

```python
# 15km 이상을 달릴 확률은?
mu = 12 #자동자차의 1리터당 평균 주행거리
sigma = 3 #표준편차 
x1 =15 
prob2 = stats.norm.cdf(x1, mu, sigma) # -무한대에서 15km까지의 cdf 값
#print("{0}km 까지의 cdf의 값은 {1:.3f}에 해당함".format(x1, prob2))
print("{0}km 이상을 달릴 확률은 {1:.3f}에 해당함".format(x1, (1-prob2)))
```

## 카이제곱(x^2) 분포
- 크기가 n 표본을 무작위로 반복하여 추출할 때,
- 각 표본에 대해 구한 표본분산들은 카이제곱 분포를 따름
- 여러 집단 간의 독립성이나 동질성 검정할 때 쓰인다
- 모집단의 분산을 추정하는데 쓰임
- 빈도 기반의 분포나 형태 적합도를 검정한다
- 0으로 시작함

```python
# seed 선택을 하여 매번 실행 시 동일한 값이 나오도록 설정
np.random.seed(seed=1234)
# 자유도 40, 데이터 수 1000개의 x^2분포를 따르는 데이터 생성
chisq_df40 = np.random.chisquare(df=40, size=1000)
# histogram, fit 정규분포 선 생성, kde: x^2 분포의 kde 생성 안 함
sns.distplot(chisq_df40, fit=stats.norm, kde=False)
```

```python
#np.random.seed(seed=1234)
chisq_df3 = np.random.chisquare(df=3, size=1000)
# histogram, fit 정규분포 선 생성, kde: x^2 분포의 kde 생성
sns.distplot(chisq_df3, fit=stats.norm, kde=True)
```

```python
np.random.seed(seed=1234)
chisq_df40 = np.random.chisquare(df=50, size=1000)
sns.distplot(chisq_df40, fit=stats.norm, kde=True)
```

```python
# 확률변수 x^2는 자유도가 30인 x^2 분포를 따른다. x^2 값이 10일 때, P(X<= x^2)을 계산하시오
chisq = 10
df = 30
# 누적확률 값 계산
prob = stats.chi2.cdf(chisq, df)
print("P(X<={0}):{1:.4f}".format(chisq, prob))
```

## F 분포
- 분산이 같은 두 정규모집단으로부터 크기 n1과 크기 n2의 확률 표본을 반복하여 독립적으로 추출한 후, 
- 구한 두 표본분산의 비율들의 표본분포

```python
# 확률 변수 F는 각각 자유도가 15, 15인 F분포를 따른다. F값이 2.0일 때, P(X <= F)을 계산
f=2.0
dfnum=15
dfden=15
prob = stats.f.cdf(x=f, dfn=dfnum, dfd=dfden)
print("P(X<=F):{:.3f}".format(prob))
```

```python
f=1.5
dfnum=10
dfden=100
prob = stats.f.cdf(x=f, dfn=dfnum, dfd=dfden)
print("P(X<=F):{:.3f}".format(prob))
```

## 포아송 분포
- 일정한 시간 또는 일정한 공간에서 발생하는 성공횟수에 대한 이산확률분포

- 어느 전화 교환대에는 1분당 평균 2회의 전화가 걸려 온다고 한다
- 전화의 도착 횟수가 포아송 분포를 따른다면,
 - 이 교환대에 1분당 3번의 전화가 걸려올 확률은 얼마일까요?
 - 이 교환대에 1분당 최대 2회 이하의 전화가 걸려올 확률은 얼마일까요?

```python
mu = 2
prob = stats.poisson.pmf(3, mu)
cdf_prob = stats.poisson.cdf(2, mu)
print("1분당 {0}번의 전화가 걸려올 확률: {1:.1f}".format(3, prob*100))
print("1분당 최대 {0}회 이하의 전화가 걸려올 확률: {1:.1f}".format(2, cdf_prob*100))
```

- 한 상점에는 평균 1시간에 3명씩 포아송분포로 손님이 도착한다.
- 이 때 10시에서 11시 사이에 다음이 발생할 확률
    - 2명의 손님이 도착
    - 1명 이하의 손님이 도착
    - 3명 이상의 손님이 도착

```python
mu = 3
prob_2 = stats.poisson.pmf(2, mu) #1) 2명의 손님이 도착
prob_1 = stats.poisson.cdf(1, mu) #2) 1명 이하의 손님이 도착
prob_3 = 1 - (stats.poisson.cdf(2, mu)) #3) 3명 이상의 손님이 도착
print("2명의 손님이 도착할 확률은 {0:.3f}이며,\n 1명 이하의 손님이 도착할 확률은 {1:.3f}이고,\n 3명 이상의 손님이 도착할 확률은 총합 1에서 2명 이하의 손님이 도착할 확률을 뺀 {2:.3f}이다.\n".format(prob_2, prob_1, prob_3))
```

# 모평균 추정
- 모표준편차를 아는 경우

- 새로운 품종의 배추 40통을 랜덤 추출하여 무게를 측정한 결과가 다음과 같음
- 모표준편차는 0.397
- 신품종 배추의 평균 무게에 대한 95%의 신뢰구간을 구하시오

```python
import pandas as pd
df = pd.DataFrame({'sample': [3.6,2.9,2.8,2.6,3.4,3.2,2.2,2.6,2.6,2.4,2.4,2.6,3.4,2.4,2.1,2.6,3.0,2.4,2.7,2.4,
                              2.6,2.9,2.9,2.0,2.7,2.2,2.8,2.7,1.8,2.5,3.0,3.2,2.8,2.6,3.2,3.1,2.9,2.7,2.7,2.2]})
df.head()
```

```python
df.isna().sum()
```

```python
lower, upper = stats.norm.interval(0.95, loc=np.mean(df), scale=0.397/np.sqrt(40))
print("신뢰구간: ({0}, {1})".format(lower.round(4), upper.round(4)))
```

```python
lower, upper = stats.norm.interval(0.90, loc=np.mean(df), scale=0.397/np.sqrt(40))
print("신뢰구간: ({0}, {1})".format(lower.round(4), upper.round(4)))
```
